Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.

```python
def second_task(N, first, d):
  an = 0
  diag = []
  for i in range(N): # np.fill_diagonal, np.diag задает диагональную матрицу
    an = first + d
    diag.append(an)
    first = an
  A = np.eye(N)
  for i in range(N):
    A[i][i] = diag[i]
  return A

second_task(7,-10,5)
```

Постройте диаграммы рассеяния для признаков Rooms, Square, HouseFloor, HouseYear в зависимости от целевой переменной Price в одной области figure. Оцените визуально, есть ли среди них такие, на которых разброс точек близок к линейной функции.

При помощи сетки графиков PairGrid визуализируйте попарные отношения признаков Rooms, Square, HouseFloor, HouseYear, Price следующим образом: на диагонали - гистограммы, под диагональю - ядерные оценки плотности, над диагональю - диаграммы рассеяния. По результатам визуализации сделайте выводы.

```python
import seaborn as sns

pg = sns.PairGrid(original, vars=['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price'])
pg.map_diag(plt.hist) # на диагонали - гистограммы
pg.map_lower(sns.kdeplot) #под диагональю - ядерные оценки плотности
pg.map_upper(plt.scatter) #над диагональю - диаграммы рассеяния
```

1. Количество комнат (Rooms) и площадь (Square) имеют положительную корреляцию, что ожидаемо.
2. Цена (Price) также имеет положительную корреляцию с площадью (Square), что также логично.
3. Высота дома (HouseFloor) и год постройки (HouseYear) не имеют сильной связи с другими признаками.
4. На диаграммах рассеяния видно, что цена (Price) имеет выбросы в больших значениях, а также есть некоторая линейная зависимость между площадью (Square) и ценой (Price).
